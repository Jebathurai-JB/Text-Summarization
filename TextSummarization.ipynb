```python
#  IMPORTING AND LOADING THE PRETRAINED MODEL:

# Importing AutoTokenizer and AutoModelForSeq2SeqLM from Huggingface transformers
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM

# AutoTokenizer for coverting the sentence into numbers 
tokenizer = AutoTokenizer.from_pretrained("google/pegasus-xsum")

# AutoModelForSeq2SeqLM for Pegasus Model
model = AutoModelForSeq2SeqLM.from_pretrained("google/pegasus-xsum")
```

```python
# FC Barcelona(Football club) transfer rumours from random blog
text = """

Alemany, who travelled to Morocco without his second in command, Jordi Cruyff, has openly expressed Barcelona’s wish to continue with Dembele. 
Although he still believes the player needs to make some compromises in order to land a new contract. At the same time, Dembele is adamant about 
getting a contract that would take him on level terms with what Samuel Umtiti and Antoine Griezmann were pocketing from their previous contracts.
It has been learned that Dembele hasn’t really signed any pre-contract agreement with a foreign team. But Sissoko did remind Barcelona that he has 
no shortage of quality suitors. And so the narrative goes on. It will take more than a couple of meetings before a final verdict is delivered. 
But one can rightly expect Barcelona to take a decision that is well within their financial framework. Meanwhile, the 24-year-old has been putting 
in extra shifts lately since he believes he is in the best shape of his Barcelona career. On Tuesday, the winger joined Ronald Araujo, Sergio Busquets, 
and Clement Lenglet to do the recovery treatment as well as regular training at the Ciutat Esportiva.

"""
```

```python
# Creating tokens: number representation of our text
tokens = tokenizer(text , truncation=True, padding="longest", return_tensors="pt")

tokens
```

```python
# Generating summary of the text(using the tokens)
summary = model.generate(**tokens)

# Summary in the form of tokens
summary
  
```

```python
# Decoding the summary to text 
tokenizer.decode(summary[0])
```
